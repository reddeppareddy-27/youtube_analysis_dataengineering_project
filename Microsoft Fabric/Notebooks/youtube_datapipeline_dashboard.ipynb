```python
from pyspark.sql import functions as F
from pyspark.sql.window import Window

df = spark.table("youtube_analytics")
df.show(5)
```

```python
df.count()
```

```python
df.select("title", "views", "likes").show(10)
```

```python
df.filter(F.col("views") > 100000).show(10)
```

```python
df.groupBy("category_id") \
  .agg(F.count("*").alias("total_videos")) \
  .orderBy(F.desc("total_videos")) \
  .show()
```

```python
df.groupBy("region") \
  .agg(F.count("*").alias("total_videos")) \
  .orderBy(F.desc("total_videos")) \
  .show()
```

```python
df.orderBy(F.desc("views")) \
  .select("title", "views") \
  .show(10)
```

```python
df = df.withColumn("engagement", F.col("likes") + F.col("comment_count"))
df.select("title", "engagement").show(10)
```

```python
df.orderBy(F.desc("engagement")) \
  .select("title", "engagement") \
  .show(10)
```

```python
df.groupBy("category_id") \
  .agg(F.avg("views").alias("avg_views")) \
  .orderBy(F.desc("avg_views")) \
  .show()
```

```python
df.groupBy("channel_title") \
  .agg(F.sum("views").alias("total_views")) \
  .orderBy(F.desc("total_views")) \
  .show(10)
```

```python
df.withColumn("engagement_rate", F.col("likes") / F.col("views")) \
  .select("title", "engagement_rate") \
  .orderBy(F.desc("engagement_rate")) \
  .show(10)
```

```python
df.groupBy("trending_date") \
  .agg(F.count("*").alias("videos_per_day")) \
  .orderBy("trending_date") \
  .show()
```

```python
window_spec = Window.partitionBy("region").orderBy(F.desc("views"))

df.withColumn("rank", F.row_number().over(window_spec)) \
  .filter(F.col("rank") <= 5) \
  .select("region", "title", "views", "rank") \
  .show()
```

```python
cat_df = df.groupBy("category_id") \
           .agg(F.avg("likes").alias("avg_likes"))

window_spec = Window.orderBy(F.desc("avg_likes"))

cat_df.withColumn("rank", F.rank().over(window_spec)).show()
```

```python
df.groupBy("region") \
  .agg(
      F.avg("views").alias("avg_views"),
      F.avg("likes").alias("avg_likes"),
      F.avg("comment_count").alias("avg_comments")
  ).orderBy(F.desc("avg_views")).show()
```

```python
df.withColumn("total_engagement", 
              F.col("views") + F.col("likes") + F.col("comment_count")) \
  .orderBy(F.desc("total_engagement")) \
  .select("title", "total_engagement") \
  .show(10)
```

```python
avg_views = df.agg(F.avg("views")).collect()[0][0]

df.filter(F.col("views") > avg_views) \
  .select("title", "views") \
  .show()
```

```python
df.groupBy("category_id") \
  .agg(
      F.sum("likes").alias("total_likes"),
      F.sum("comment_count").alias("total_comments")
  ).orderBy(F.desc("total_likes")).show()
```

```python
df.groupBy("channel_title") \
  .agg(F.count("*").alias("total_videos")) \
  .orderBy(F.desc("total_videos")) \
  .show(10)
```

```python
region_cat = df.groupBy("region", "category_id") \
               .agg(F.count("*").alias("total_videos"))

window_spec = Window.partitionBy("region").orderBy(F.desc("total_videos"))

region_cat.withColumn("rank", F.rank().over(window_spec)) \
          .filter(F.col("rank") == 1) \
          .show()
```

```python
%%sql
SELECT * FROM youtube LIMIT 10;
```

```python
%%sql
SELECT COUNT(*) AS total_records FROM youtube;
```

```python
%%sql
SELECT 
    category_id,
    COUNT(*) AS total_videos
FROM youtube
GROUP BY category_id
ORDER BY total_videos DESC;
```

```python
%%sql
SELECT 
    region,
    COUNT(*) AS total_videos
FROM youtube
GROUP BY region
ORDER BY total_videos DESC;
```

```python
%%sql
SELECT 
    title,
    views
FROM youtube
ORDER BY views DESC
LIMIT 10;
```

```python
%%sql
SELECT 
    title,
    views,
    likes,
    comment_count,
    (likes + comment_count) AS engagement
FROM youtube
ORDER BY engagement DESC
LIMIT 10;
```

```python
%%sql
SELECT 
    category_id,
    AVG(views) AS avg_views
FROM youtube
GROUP BY category_id
ORDER BY avg_views DESC;
```

```python
%%sql
SELECT 
    channel_title,
    SUM(views) AS total_views
FROM youtube
GROUP BY channel_title
ORDER BY total_views DESC
LIMIT 10;
```

```python
%%sql
SELECT 
    title,
    views,
    likes,
    ROUND(likes / views, 4) AS engagement_rate
FROM youtube
WHERE views > 100000
ORDER BY engagement_rate DESC
LIMIT 10;
```

```python
%%sql
SELECT 
    trending_date,
    COUNT(*) AS videos_per_day
FROM youtube
GROUP BY trending_date
ORDER BY trending_date;
```

```python
%%sql
SELECT *
FROM (
    SELECT 
        region,
        title,
        views,
        ROW_NUMBER() OVER (PARTITION BY region ORDER BY views DESC) AS rank
    FROM youtube
) t
WHERE rank <= 5;
```

```python
%%sql
SELECT *
FROM (
    SELECT 
        category_id,
        AVG(likes) AS avg_likes,
        RANK() OVER (ORDER BY AVG(likes) DESC) AS rank
    FROM youtube
    GROUP BY category_id
) t;
```

```python
%%sql
SELECT 
    region,
    AVG(views) AS avg_views,
    AVG(likes) AS avg_likes,
    AVG(comment_count) AS avg_comments
FROM youtube
GROUP BY region
ORDER BY avg_views DESC;
```

```python
%%sql
SELECT 
    title,
    views,
    likes,
    comment_count,
    (views + likes + comment_count) AS total_engagement
FROM youtube
ORDER BY total_engagement DESC
LIMIT 10;
```

```python
%%sql
SELECT 
    title,
    views
FROM youtube
WHERE views > (
    SELECT AVG(views) FROM youtube
);
```

```python
%%sql
SELECT 
    category_id,
    SUM(likes) AS total_likes,
    SUM(comment_count) AS total_comments
FROM youtube
GROUP BY category_id
ORDER BY total_likes DESC;
```

```python
%%sql
SELECT 
    channel_title,
    COUNT(*) AS total_videos
FROM youtube
GROUP BY channel_title
ORDER BY total_videos DESC
LIMIT 10;
```

```python
%%sql
SELECT 
    region,
    SUM(views) AS total_views
FROM youtube
GROUP BY region
ORDER BY total_views DESC;
```

```python
%%sql
SELECT 
    category_id,
    AVG(likes / views) AS like_view_ratio
FROM youtube
GROUP BY category_id
ORDER BY like_view_ratio DESC;
```

```python
%%sql
SELECT *
FROM (
    SELECT 
        region,
        category_id,
        COUNT(*) AS total_videos,
        RANK() OVER (PARTITION BY region ORDER BY COUNT(*) DESC) AS rank
    FROM youtube
    GROUP BY region, category_id
) t
WHERE rank = 1;
```
